```python
from dotenv import load_dotenv
import os
load_dotenv()
api_key = os.getenv("HUGGING_FACE_KEY")
```

```python
import langchain
from langchain.llms import HuggingFacePipeline
```

```python
from transformers import AutoTokenizer
```

```python
import transformers
```

```python
import torch
```

```python
import warnings
warnings.filterwarnings("ignore")
```

# Loading the model

```python
# model="meta-llama/Llama-2-7b-chat-hf"
model = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
```

```python
tokenizer=AutoTokenizer.from_pretrained(model,token = api_key)
```

```python
pipeline=transformers.pipeline(
    "text-generation",
    model=model,
    token= api_key,
    tokenizer=tokenizer,
    dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=1000,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
    )
```

```python
llm = HuggingFacePipeline(pipeline =pipeline , model_kwargs={'temperature':0} )
```

```python
prmpt= "what is the good name for a company that makes colorful shirts and googles "
```

```python
print(llm(prmpt))
```

```python
from langchain.chains import LLMChain
```

```python
from langchain.prompts import PromptTemplate
```

```python
prompt_template2=PromptTemplate(input_variables=["book_name"],
                               template="Provide me a concise summary of the book {book_name}")
```

```python
input_prompt=prompt_template2.format(book_name="Alchemist")
```

```python
print(input_prompt)
```

```python
chain = LLMChain(llm=llm, prompt=prompt_template2, verbose=True)
response= chain.run("Harry Potter")
print(response)
```
